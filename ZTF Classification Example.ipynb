# Section 2: Classification

```python
#This next section will show how to process the training data and train the classifier
SupernovaTrainingData = pd.read_csv(r"C:/Users/blgnm/ZTF/ClassifierTraining.csv",index_col = 0)

#BogusTrainingData = pd.read_csv(r"C:/Users/blgnm/Supernova Classification/Bogus.csv", index_col = 0)
#Note: I reccomend finding a more representative Bogus Training set. I've found the one I have isn't as representitive as it
#should be.
```

```python
ZTFData = SupernovaClassification.ZTFData

#Classified Supernova Training Data
SupernovaFunctionParam, SupernovaCoeff = ZTFData().Wavelet2(SupernovaTrainingData,WaveletType = 'sym2', classification=True, Date = 100, length = 100000)

#Note I personally prefer to not exclude any data out of the trainingset, so I set max event length to a very high number


#Bogus Training Data
#BogusFunctionParam, BogusCoeff = ZTFData().Wavelet(BogusTrainingData,WaveletType = 'sym5', classification=True)
#Note: it takes a lot longer to process the Bogus Data
```

```python
#Reduce the dimensions using PCA
#Change smot argument to choose whether or not to apply SMOTE (applies before pca)
ProcessedData, PCA = ZTFData().DimensionalityReduction2(SupernovaCoeff, SupernovaFunctionParam, n = 3, smot = True, Trainset = True)

### Very Important
#Choosing to use SMOTE or not to use SMOTE on the training data will change the order of columns in the data frame since event
#names are no longer included. Keep this in mind when inputting into the classifier so that you don't get an error
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import NearestNeighbors
svc = RandomForestClassifier(n_estimators = 30, min_samples_split = 6, max_features = 10)
Classifier = ZTFData().SupernovaTypeClassifierTrainer(Train = ProcessedData.iloc[:,1:], y =  ProcessedData['class'], KNN = True, Ada = False, p =1, n_neighbors =11)
```

```python
#Now we load in the testing Data Frame REFITT (Querried with the following requirements: min alerts = 30, filter: REFITT)
#You can set the length of events included manualy (Doing this will filter out most bogus data)
TestingData = pd.read_csv(r"C://Users/blgnm/ZTF/ClassifiedREFITTFilter.csv",index_col = 0)
```

```python
#computing the wavelet tronsform on the training data also takes a while
TestingParam, TestingCoeff = ZTFData().Wavelet2(TestingData, Date = 100, length = 1500)
#Note: if you notice that a progress bar gets completed before hitting 100%, that is because of the length filter
```

```python
#Note: Pay attention to the Trainset Parameter, when set to false, pca is not applied, it just formats the data so you can
#call the already fitted pca from when you performed it on the training data.
ProcessedTrainingData = ZTFData().DimensionalityReduction2(TestingCoeff, TestingParam, n = 3, Trainset = False)

x = PCA.transform(ProcessedTrainingData.iloc[:,2:])
ProcessedTraining = pd.concat([ProcessedTrainingData.iloc[:,:2].reset_index(drop=True), pd.DataFrame(x)], axis=1)
print(ProcessedTraining)

#Note: if you get a training data dimensions must match number of dimensions at time of training data, check to see
# if you changed your indexing when changing the smot parameter in Dimensionality reduction
```

```python
labels = pd.DataFrame(Classifier.predict(ProcessedTraining.iloc[:,2:]))
proba = pd.DataFrame(Classifier.predict_proba(ProcessedTraining.iloc[:,2:]), columns = ['1a', '2', '1bc', 'SLSN'])

LabeledData = pd.concat([labels.reset_index(drop=True), proba.reset_index(drop=True),ProcessedTraining.reset_index(drop=True) ],axis=1)

print(LabeledData.iloc[:,0].value_counts())
```
